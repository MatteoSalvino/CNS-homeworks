```python
from cryptography.hazmat.primitives.ciphers import Cipher, algorithms, modes
from cryptography.hazmat.backends import default_backend
import os
import base64
```

```python
key=b"1708108"
file = open('./sample_image.jpeg', 'rb')
msg = file.read()
file.close()
iv = os.urandom(8)
algorithm = algorithms.Blowfish(key)
cipher = Cipher(algorithm, mode=modes.CBC(iv), backend=default_backend())
encryptor = cipher.encryptor()
ct = encryptor.update(msg)
print(ct)
```

```python
from Crypto.Cipher import Blowfish
```

```python
cipher2 = Blowfish.new(key, Blowfish.MODE_CBC, iv)
pt = cipher2.decrypt(ct)
print(pt)
file = open('./res.jpeg', 'wb')
file.write(pt)
file.close()
```
